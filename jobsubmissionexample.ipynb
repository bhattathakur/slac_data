```python
#mac template
#/control/saveHistory
import random
import os
isoza={'am241':[95,241],'ba133':[56,133],'co60':[27,60],'cs137':[55,137],'eu152':[63,152],'na22':[11,22],'pb210':[82,210],'ra226':[88,226]}

#/gps/pos/centre 277.9 0 0 mm  #for door
#/gps/pos/centre 266.9 -139.9 -150.9 mm  #front left corner
#/gps/pos/centre -8.358000 -139.9 -150.9 mm  #corner
pos_cordinate={'door':'277.9 0 0','corner':'-8.358000 -139.9 -150.9','frontleftcorner':'266.9 -139.9 -150.9'}
pos_cordinate_ra226={'door':'276.4 0 0','corner':'-8.358000 -139.9 -150.9','frontleftcorner':'266.9 -139.9 -150.9'} #need to check for corner, frontleftcorner for ra226

position='frontleftcorner' #door, corner
isotope='am241'
def get_mac_file(isotope,position,seed):
    print(f'seed   :{seed}')
    print(f'isotope:{isotope}')
    mac_file=f'{isotope}{seed}.mac'
    print(f"mac_file: {mac_file}\n")
    part1= f"""
/gesim/phys/SelectPhysics QGSP_BIC_HP
#/gesim/phys/SelectPhysics Hadron
#BEGIN savegen
/gesim/analysis/saveGen no
#/gesim/analysis/saveGen yes
#END savegen

# Material definitions has to be done before /run/initialize
#BEGIN detselect
/gesim/detector/selectDetector 4
/gesim/analysis/selectDetector 4
/gesim/source/setSourcePosition {position}_{isotope}
#END detselect

/run/initialize
#/grdm/analogueMC 1
/grdm/verbose 0
/grdm/allVolumes
"""
    if isotope in ['am241']:
        part2=f"""
/grdm/nucleusLimits 240 241 94 95 #for am
"""
    elif isotope in ['pb210']: 
        part2= f"""
/grdm/nucleusLimits 210 210 82 82 #for pb
"""
    else:
        part2=f"""
"""
#/grdm/setPhotoEvaporationFile 78 196 raddata/evap5_z78.a196
#/grdm/setRadioactiveDecayFile 80 200 raddata/rad5_z80.a200
#/grdm/selectVolume Target

# Next lines needed to have atomic relaxation working 
    part3=f"""
/process/em/fluo true
/process/em/auger true


# Print primary events (yes/no)
#/event/printEvents yes
/event/printEvents no

# Simulate 
#/gps/verbose 2
/histo/fileName {isotope}_{position}{seed} #rootfile to save events
#BEGIN seed
/random/setSeeds {random.randint(1,1000000000)}  {random.randint(1,100000)}
#END seed

#BEGIN evtgen
/gps/particle ion
/gps/ion {isoza[isotope][0]} {isoza[isotope][1]} 0 0
#END evtgen

#BEGIN srcselect
/gps/ang/type iso
/gps/pos/type Volume
#/gps/pos/shape Sphere
/gps/pos/shape Para
"""
    if isotope not in ['ra226']:
        part4=f"""
/gps/pos/centre {pos_cordinate[position]} mm  #for {position}
/gps/pos/halfx  0.5 mm
/gps/pos/halfy  0.5 mm
/gps/pos/halfz  0.5 mm
/gps/pos/confine Bead
"""
    else:
        part4=f"""
/gps/pos/centre {pos_cordinate_ra226[position]} mm  #for {position}
/gps/pos/halfx  0.15 mm
/gps/pos/halfy  1.2 mm
/gps/pos/halfz  1.2 mm
/gps/pos/confine Foil
"""
#/gps/pos/confine InnerCavity
#END srcselect
#/gps/source/show #show the source
    part5=f"""
/run/verbose 1
/run/printProgress 100000
#BEGIN nevts
/run/beamOn 100000
"""
    mac_script=part1+part2+part3+part4+part5
    macfile = open( mac_file, 'w' )
    macfile.write( mac_script )
    macfile.close()
    return part1+part2+part3+part4+part5
```

```python


def bash_file(isotope,seed):
    print(f'seed   :{seed}')
    print(f'isotope:{isotope}')
    batch_file=f'{isotope}{seed}.sh'
    print(f"batch_file: {batch_file}\n")
    bash_script=f"""
#bash template
#!/bin/bash
#SBATCH --time=3-00:00:00   # Time limit for the job (REQUIRED).
#SBATCH --job-name=ra2263   # Job name
##SBATCH --gres=gpu:1      #number of gpu per task
#SBATCH --nodes=1           # number of nodes
##SBATCH --ntasks=1          # Number of cores for the job. Same as SBATCH -n 1
##SBATCH --mem=32g           #32 GB ram asked
##SBATCH --partition=CAL48M192_L #choose one of the nodes
##SBATCH --partition=CAC48M192_L #choose one of the nodes
#SBATCH --partition=SKY32M192_L
##SBATCH --partition=P4V16_HAS16M128_L
##SBATCH --partition=P4V12_SKY32M192_L
#SBATCH -e {isotope}{seed}-%j.err
#SBATCH -o {isotope}{seed}-%j.log
#SBATCH -A col_rma422_uksr  # Project allocation account name (REQUIRED)
#SBATCH --mail-type FAIL    # Send email when job starts/ends
#SBATCH --mail-user tpbh222@uky.edu# Where email is sent to (optional)


echo "Job $SLURM_JOB_ID running on SLURM NODELIST: $SLURM_NODELIST "
echo "`date`"

#loading the essential modules
module purge
module load cmake/3.14.3
module load ccs/conda/root-6.22.0+geant4-10.6.0 
$HOME/gesim/gesim {isotope}{seed}.mac          #gesim
echo "DONE!"
echo "`date`"
"""
    bashfile = open( batch_file, 'w' )
    bashfile.write( bash_script )
    bashfile.close()
    os.system( "sbatch " + batchfile)
    return bash_script
```
